```python
#Multiple Length Arguments----->*args,**kwargs
```

```python
def hello(*args):
    print(f'args---->{args}')
    print(type(args))
```

```python
hello(1,2,3,4,4,5,5,6,7,10,8,9,)
```

```python
def hi(*args):
    for i in args:
        print(i)
```

```python
hi(1,2,'hey','hello',1)
```

```python
def hello(**kwargs):
    print(kwargs)
    print(type(kwargs))
```

```python
hello(1,2)
```

```python
hello(Name='Sachin',Course='DS')
```

```python
def hi(**abc):
    for i,j in abc.items():
        print(f'{i}--->{j}')
```

```python
hi(Name='Sachin',Course='DS')
```

```python
def isLower(x):
    for i in x:
        if ord(i)>122 or ord(i)<97:
            return False
    else:
        return True
```

```python
isLower('sachin')
```

```python
isLower('sachIn')
```

```python
%%writefile primemodule.py
def prime(num):
    for i in range(2,num//2+1):
        if num%i==0:
            return False
    else:
        return True
```

```python
import primemodule
```

```python
primemodule.prime(121)
```

```python
primemodule.prime(127)
```
